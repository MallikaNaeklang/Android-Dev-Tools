```python
##Develope code by Kapomtiiz SB MD KKU##

import pandas as pd

# Load the dataset with better handling of mixed data types
file_path = "mzmine_pos_HMDB_GNPS_Agilent.csv"
df = pd.read_csv(file_path, low_memory=False)

# Displays all column names (df.columns.tolist()) to verify if they match the expected ones.
print(df.columns.tolist())

# Uses df.columns.str.strip() to remove any leading or trailing spaces from column names.
df.columns = df.columns.str.strip()

# Define the required columns
required_columns = [
    "id", "rt", "rt_range:min", "rt_range:max", "mz", "mz_range:min", "mz_range:max", "height", "area",
    "intensity_range:min", "intensity_range:max", "fragment_scans", "charge", "feature_group",
    "ion_identities:iin_id", "ion_identities:ion_identities", "ion_identities:list_size",
    "ion_identities:neutral_mass", "ion_identities:partner_row_ids", "ion_identities:iin_relationship",
    "ion_identities:consensus_formulas", "ion_identities:simple_formulas",
    "spectral_db_matches:spectral_db_matches", "spectral_db_matches:compound_name",
    "spectral_db_matches:similarity_score", "spectral_db_matches:n_matching_signals"
]

# Include any column that contains "area"
area_columns = [col for col in df.columns if "area" in col.lower() and col not in required_columns]

# Ensures only existing columns are selected (existing_columns) to prevent KeyError.
existing_columns = [col for col in required_columns if col in df.columns]
selected_columns = existing_columns + area_columns

# Filter the dataframe with selected columns
df_filtered = df[selected_columns]

# Replace empty values in columns containing "area" with 0
df_filtered[area_columns] = df_filtered[area_columns].fillna(0)

# Save the cleaned dataframe to a new CSV file
output_file = "cleaned_ChP_pos_data.csv"
df_filtered.to_csv(output_file, index=False)

print(f"Cleaned data has been saved to {output_file}")
```
